````python
# Complete yfinance Tutorial - Python Financial Data Library

## Table of Contents
1. [Installation](#installation)
2. [Basic Usage](#basic-usage)
3. [Stock Data Retrieval](#stock-data-retrieval)
4. [Historical Data](#historical-data)
5. [Company Information](#company-information)
6. [Financial Statements](#financial-statements)
7. [Multiple Stocks](#multiple-stocks)
8. [Real-time Data](#real-time-data)
9. [Options Data](#options-data)
10. [Market Indices](#market-indices)
11. [Data Visualization](#data-visualization)
12. [Advanced Features](#advanced-features)
13. [Best Practices](#best-practices)

## Installation

```bash
pip install yfinance
```

For additional features with plotting:
```bash
pip install yfinance pandas matplotlib plotly
```

## Basic Usage

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Create a Ticker object
ticker = yf.Ticker("AAPL")
```

## Stock Data Retrieval

### Current Stock Info
```python
# Get stock info
info = ticker.info
print(f"Company: {info['longName']}")
print(f"Current Price: ${info['currentPrice']}")
print(f"Market Cap: ${info['marketCap']:,}")
print(f"P/E Ratio: {info.get('trailingPE', 'N/A')}")
```

### Basic Stock Data
```python
# Get current stock data
data = yf.download("AAPL", period="1d", interval="1m")
print(data.tail())
```

## Historical Data

### Different Time Periods
```python
# Valid periods: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max
periods = ["1mo", "3mo", "6mo", "1y", "2y"]

for period in periods:
    data = yf.download("AAPL", period=period)
    print(f"\n{period.upper()} Data:")
    print(f"Start: {data.index[0].strftime('%Y-%m-%d')}")
    print(f"End: {data.index[-1].strftime('%Y-%m-%d')}")
    print(f"Total Days: {len(data)}")
```

### Custom Date Range
```python
# Download data for specific date range
data = yf.download("AAPL", start="2023-01-01", end="2024-01-01")
print(data.head())

# With specific interval
# Valid intervals: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
data_hourly = yf.download("AAPL", period="5d", interval="1h")
```

### Data Structure Understanding
```python
# Understanding the data structure
data = yf.download("AAPL", period="1mo")
print("Columns:", data.columns.tolist())
print("Index type:", type(data.index))
print("Data types:")
print(data.dtypes)

# Accessing specific columns
print("\nClosing prices:")
print(data['Close'].head())

# Calculate daily returns
data['Daily_Return'] = data['Close'].pct_change()
print("\nDaily returns:")
print(data['Daily_Return'].head())
```

## Company Information

### Comprehensive Company Data
```python
ticker = yf.Ticker("AAPL")

# Basic company information
info = ticker.info

# Key metrics
key_metrics = {
    'Company Name': info.get('longName'),
    'Sector': info.get('sector'),
    'Industry': info.get('industry'),
    'Country': info.get('country'),
    'Website': info.get('website'),
    'Business Summary': info.get('longBusinessSummary', '')[:200] + '...',
    'Market Cap': info.get('marketCap'),
    'Enterprise Value': info.get('enterpriseValue'),
    'P/E Ratio': info.get('trailingPE'),
    'Forward P/E': info.get('forwardPE'),
    'PEG Ratio': info.get('pegRatio'),
    'Price to Book': info.get('priceToBook'),
    'Debt to Equity': info.get('debtToEquity'),
    'ROE': info.get('returnOnEquity'),
    'ROA': info.get('returnOnAssets'),
    'Dividend Yield': info.get('dividendYield'),
    '52 Week High': info.get('fiftyTwoWeekHigh'),
    '52 Week Low': info.get('fiftyTwoWeekLow'),
    'Beta': info.get('beta'),
    'Volume': info.get('volume'),
    'Average Volume': info.get('averageVolume')
}

for key, value in key_metrics.items():
    if value is not None:
        print(f"{key}: {value}")
```

## Financial Statements

### Income Statement
```python
ticker = yf.Ticker("AAPL")

# Get income statement (annual)
income_stmt = ticker.income_stmt
print("Annual Income Statement:")
print(income_stmt.head())

# Quarterly income statement
quarterly_income = ticker.quarterly_income_stmt
print("\nQuarterly Income Statement:")
print(quarterly_income.head())
```

### Balance Sheet
```python
# Annual balance sheet
balance_sheet = ticker.balance_sheet
print("Annual Balance Sheet:")
print(balance_sheet.head())

# Quarterly balance sheet
quarterly_balance = ticker.quarterly_balance_sheet
print("\nQuarterly Balance Sheet:")
print(quarterly_balance.head())
```

### Cash Flow Statement
```python
# Annual cash flow
cash_flow = ticker.cash_flow
print("Annual Cash Flow:")
print(cash_flow.head())

# Quarterly cash flow
quarterly_cash_flow = ticker.quarterly_cash_flow
```

### Financial Analysis Example
```python
# Calculate some financial ratios
def analyze_financials(ticker_symbol):
    ticker = yf.Ticker(ticker_symbol)
    info = ticker.info
    
    # Get latest financial data
    income_stmt = ticker.income_stmt
    balance_sheet = ticker.balance_sheet
    
    if not income_stmt.empty and not balance_sheet.empty:
        latest_year = income_stmt.columns[0]
        
        # Revenue growth (if multiple years available)
        if len(income_stmt.columns) > 1:
            current_revenue = income_stmt.loc['Total Revenue', income_stmt.columns[0]]
            previous_revenue = income_stmt.loc['Total Revenue', income_stmt.columns[1]]
            revenue_growth = ((current_revenue - previous_revenue) / previous_revenue) * 100
            print(f"Revenue Growth: {revenue_growth:.2f}%")
        
        # Profit margin
        net_income = income_stmt.loc['Net Income', latest_year]
        total_revenue = income_stmt.loc['Total Revenue', latest_year]
        profit_margin = (net_income / total_revenue) * 100
        print(f"Profit Margin: {profit_margin:.2f}%")
        
        # Current ratio
        if 'Current Assets' in balance_sheet.index and 'Current Liabilities' in balance_sheet.index:
            current_assets = balance_sheet.loc['Current Assets', latest_year]
            current_liabilities = balance_sheet.loc['Current Liabilities', latest_year]
            current_ratio = current_assets / current_liabilities
            print(f"Current Ratio: {current_ratio:.2f}")

analyze_financials("AAPL")
```

## Multiple Stocks

### Download Multiple Stocks
```python
# Multiple stocks at once
tickers = ["AAPL", "GOOGL", "MSFT", "TSLA", "AMZN"]
data = yf.download(tickers, period="1y")

print("Shape of data:", data.shape)
print("Stocks:", tickers)

# Access specific stock data
apple_close = data['Close']['AAPL']
print("\nApple closing prices (last 5 days):")
print(apple_close.tail())
```

### Portfolio Analysis
```python
# Create a simple portfolio analysis
def portfolio_analysis(tickers, period="1y"):
    data = yf.download(tickers, period=period)['Close']
    
    # Calculate daily returns
    returns = data.pct_change().dropna()
    
    # Portfolio metrics
    print("=== Portfolio Analysis ===")
    print(f"Period: {period}")
    print(f"Stocks: {', '.join(tickers)}")
    print("\nAnnualized Returns:")
    annual_returns = returns.mean() * 252
    for ticker in tickers:
        print(f"{ticker}: {annual_returns[ticker]:.2%}")
    
    print("\nVolatility (Annual):")
    volatility = returns.std() * (252**0.5)
    for ticker in tickers:
        print(f"{ticker}: {volatility[ticker]:.2%}")
    
    print("\nSharpe Ratio (assuming 2% risk-free rate):")
    risk_free_rate = 0.02
    sharpe_ratios = (annual_returns - risk_free_rate) / volatility
    for ticker in tickers:
        print(f"{ticker}: {sharpe_ratios[ticker]:.2f}")
    
    # Correlation matrix
    print("\nCorrelation Matrix:")
    correlation = returns.corr()
    print(correlation.round(3))
    
    return data, returns

# Run portfolio analysis
portfolio_data, portfolio_returns = portfolio_analysis(["AAPL", "GOOGL", "MSFT"])
```

## Real-time Data

### Current Market Data
```python
def get_realtime_data(tickers):
    """Get real-time data for multiple tickers"""
    data = {}
    
    for ticker_symbol in tickers:
        ticker = yf.Ticker(ticker_symbol)
        info = ticker.info
        
        data[ticker_symbol] = {
            'Current Price': info.get('currentPrice', 'N/A'),
            'Previous Close': info.get('previousClose', 'N/A'),
            'Change': info.get('currentPrice', 0) - info.get('previousClose', 0),
            'Change %': ((info.get('currentPrice', 0) - info.get('previousClose', 0)) / info.get('previousClose', 1)) * 100,
            'Volume': info.get('volume', 'N/A'),
            'Market Cap': info.get('marketCap', 'N/A')
        }
    
    return data

# Get real-time data
realtime = get_realtime_data(["AAPL", "GOOGL", "MSFT"])
for ticker, data in realtime.items():
    print(f"\n{ticker}:")
    for key, value in data.items():
        if isinstance(value, float):
            print(f"  {key}: {value:.2f}")
        else:
            print(f"  {key}: {value}")
```

## Options Data

### Options Chain
```python
ticker = yf.Ticker("AAPL")

# Get options expiration dates
exp_dates = ticker.options
print("Available expiration dates:")
for date in exp_dates[:5]:  # Show first 5
    print(date)

# Get options chain for specific expiration
if exp_dates:
    opt_chain = ticker.option_chain(exp_dates[0])
    
    print(f"\nOptions for {exp_dates[0]}:")
    print("\nCalls (first 5):")
    print(opt_chain.calls[['strike', 'lastPrice', 'volume', 'impliedVolatility']].head())
    
    print("\nPuts (first 5):")
    print(opt_chain.puts[['strike', 'lastPrice', 'volume', 'impliedVolatility']].head())
```

## Market Indices

### Major Indices
```python
# Common market indices
indices = {
    "S&P 500": "^GSPC",
    "Dow Jones": "^DJI", 
    "NASDAQ": "^IXIC",
    "Russell 2000": "^RUT",
    "VIX": "^VIX"
}

print("Market Indices Performance:")
for name, symbol in indices.items():
    data = yf.download(symbol, period="1d")
    if not data.empty:
        current = data['Close'][-1]
        previous = data['Open'][-1]
        change = current - previous
        change_pct = (change / previous) * 100
        print(f"{name}: {current:.2f} ({change:+.2f}, {change_pct:+.2f}%)")
```

## Data Visualization

### Basic Plotting
```python
import matplotlib.pyplot as plt

# Simple price chart
data = yf.download("AAPL", period="1y")

plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'])
plt.title('Apple Stock Price - 1 Year')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()

# Multiple stocks comparison
tickers = ["AAPL", "GOOGL", "MSFT"]
data = yf.download(tickers, period="6mo")['Close']

plt.figure(figsize=(12, 6))
for ticker in tickers:
    # Normalize to show percentage change
    normalized = (data[ticker] / data[ticker].iloc[0]) * 100
    plt.plot(data.index, normalized, label=ticker)

plt.title('Stock Performance Comparison (Normalized)')
plt.xlabel('Date')
plt.ylabel('Normalized Price (%)')
plt.legend()
plt.grid(True)
plt.show()
```

### Advanced Visualization
```python
# Candlestick-style chart using matplotlib
def plot_candlestick_simple(data, ticker_name):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), 
                                   gridspec_kw={'height_ratios': [3, 1]})
    
    # Price chart
    ax1.plot(data.index, data['Close'], label='Close Price', linewidth=2)
    ax1.plot(data.index, data['High'], alpha=0.3, color='green')
    ax1.plot(data.index, data['Low'], alpha=0.3, color='red')
    ax1.fill_between(data.index, data['Low'], data['High'], alpha=0.1)
    ax1.set_title(f'{ticker_name} Stock Price')
    ax1.set_ylabel('Price ($)')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    # Volume chart
    ax2.bar(data.index, data['Volume'], alpha=0.7, color='blue')
    ax2.set_title('Volume')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Volume')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Plot Apple data
apple_data = yf.download("AAPL", period="3mo")
plot_candlestick_simple(apple_data, "Apple")
```

## Advanced Features

### Custom Analysis Functions
```python
def technical_analysis(ticker_symbol, period="1y"):
    """Perform basic technical analysis"""
    data = yf.download(ticker_symbol, period=period)
    
    # Calculate moving averages
    data['MA20'] = data['Close'].rolling(window=20).mean()
    data['MA50'] = data['Close'].rolling(window=50).mean()
    
    # Calculate RSI
    def calculate_rsi(prices, period=14):
        delta = prices.diff()
        gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
        loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
        rs = gain / loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    data['RSI'] = calculate_rsi(data['Close'])
    
    # Current signals
    current_price = data['Close'][-1]
    ma20 = data['MA20'][-1]
    ma50 = data['MA50'][-1]
    rsi = data['RSI'][-1]
    
    print(f"=== Technical Analysis for {ticker_symbol} ===")
    print(f"Current Price: ${current_price:.2f}")
    print(f"20-day MA: ${ma20:.2f}")
    print(f"50-day MA: ${ma50:.2f}")
    print(f"RSI: {rsi:.2f}")
    
    # Simple signals
    if current_price > ma20 > ma50:
        print("Signal: BULLISH (Price above both MAs)")
    elif current_price < ma20 < ma50:
        print("Signal: BEARISH (Price below both MAs)")
    else:
        print("Signal: NEUTRAL")
    
    if rsi > 70:
        print("RSI Signal: OVERBOUGHT")
    elif rsi < 30:
        print("RSI Signal: OVERSOLD")
    else:
        print("RSI Signal: NEUTRAL")
    
    return data

# Run technical analysis
tech_data = technical_analysis("AAPL")
```

### Error Handling
```python
def safe_download(ticker, **kwargs):
    """Download data with error handling"""
    try:
        data = yf.download(ticker, **kwargs)
        if data.empty:
            print(f"No data found for {ticker}")
            return None
        return data
    except Exception as e:
        print(f"Error downloading {ticker}: {str(e)}")
        return None

# Usage
data = safe_download("AAPL", period="1y")
if data is not None:
    print("Data downloaded successfully")
    print(data.tail())
```

## Best Practices

### 1. Rate Limiting and Caching
```python
import time
from functools import lru_cache

@lru_cache(maxsize=100)
def cached_download(ticker, period):
    """Cache downloaded data to avoid repeated API calls"""
    return yf.download(ticker, period=period)

def batch_download_with_delay(tickers, delay=1):
    """Download multiple tickers with delay to avoid rate limiting"""
    data = {}
    for ticker in tickers:
        print(f"Downloading {ticker}...")
        data[ticker] = yf.download(ticker, period="1y")
        time.sleep(delay)  # Add delay between requests
    return data
```

### 2. Data Validation
```python
def validate_data(data, ticker):
    """Validate downloaded data"""
    if data is None or data.empty:
        print(f"Warning: No data for {ticker}")
        return False
    
    if data.isnull().sum().sum() > 0:
        print(f"Warning: {ticker} has missing values")
        print("Missing values per column:")
        print(data.isnull().sum())
    
    if len(data) < 10:
        print(f"Warning: {ticker} has very little data ({len(data)} rows)")
    
    return True

# Usage
data = yf.download("AAPL", period="1y")
if validate_data(data, "AAPL"):
    print("Data validation passed")
```

### 3. Configuration Management
```python
# Configuration for common analysis
CONFIG = {
    'DEFAULT_PERIOD': '1y',
    'DEFAULT_INTERVAL': '1d',
    'TECH_INDICATORS': {
        'MA_SHORT': 20,
        'MA_LONG': 50,
        'RSI_PERIOD': 14
    },
    'RISK_FREE_RATE': 0.02
}

def get_stock_data(ticker, period=None, interval=None):
    """Get stock data with default configuration"""
    period = period or CONFIG['DEFAULT_PERIOD']
    interval = interval or CONFIG['DEFAULT_INTERVAL']
    return yf.download(ticker, period=period, interval=interval)
```

## Common Use Cases

### 1. Daily Market Monitor
```python
def daily_market_monitor(watchlist):
    """Monitor daily performance of watchlist"""
    print("=== Daily Market Monitor ===")
    print(f"Date: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print("-" * 50)
    
    for ticker in watchlist:
        try:
            stock = yf.Ticker(ticker)
            info = stock.info
            
            current = info.get('currentPrice', 0)
            previous = info.get('previousClose', 0)
            change = current - previous
            change_pct = (change / previous) * 100 if previous != 0 else 0
            
            print(f"{ticker:6} | ${current:8.2f} | {change:+7.2f} | {change_pct:+6.2f}%")
        except:
            print(f"{ticker:6} | Error retrieving data")

# Usage
watchlist = ["AAPL", "GOOGL", "MSFT", "TSLA", "AMZN"]
daily_market_monitor(watchlist)
```

### 2. Dividend Analysis
```python
def analyze_dividends(ticker_symbol):
    """Analyze dividend history"""
    ticker = yf.Ticker(ticker_symbol)
    
    # Get dividend data
    dividends = ticker.dividends
    if dividends.empty:
        print(f"{ticker_symbol} does not pay dividends")
        return
    
    print(f"=== Dividend Analysis for {ticker_symbol} ===")
    print(f"Total dividends paid: {len(dividends)}")
    print(f"Latest dividend: ${dividends[-1]:.3f} on {dividends.index[-1].strftime('%Y-%m-%d')}")
    
    # Annual dividend summary
    annual_dividends = dividends.groupby(dividends.index.year).sum()
    print("\nAnnual Dividends:")
    for year, amount in annual_dividends.items():
        print(f"{year}: ${amount:.3f}")
    
    # Dividend growth
    if len(annual_dividends) > 1:
        recent_growth = annual_dividends.pct_change().dropna()
        avg_growth = recent_growth.mean()
        print(f"\nAverage annual dividend growth: {avg_growth:.2%}")

# Usage
analyze_dividends("AAPL")
```

This tutorial provides a comprehensive guide to using yfinance for financial data analysis in Python. The library is powerful for both beginners and advanced users, offering access to a wide range of financial data from Yahoo Finance.
````
